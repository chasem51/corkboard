```python
import requests
import bs4
```

```python
response = requests.get('http://duspviz.mit.edu/_assets/data/county_housing_stats.html') # load and get the website w beautiful soup
```

```python
soup = bs4.BeautifulSoup(response.text, "html.parser") # create the soup
print(soup.prettify()) # print output using prettify
```

```python
soup.title
```

```python
soup.title.string # access content of the title element
```

```python
soup.p # access data w first p tag
```

```python
soup.p.string 
```

```python
soup.a.string # access data in the first a tag
```

```python
soup.find_all('a') # returns an array of all links in a document
```

```python
soup.find(id = "link3") # retreive a specific link by ID
```

```python
soup.find_all('td') # access data in the table
```

```python
data = soup.findAll(attrs={'class':'name'}) # make an array w all elements that are of the name class
```

```python
print(data[0]) # print the raw contents of the array
```

```python
for i in data:     # prints all of the array elements
    print(i.string)
```

```python
bigdata = soup.findAll(attrs={'class':['name','fips','tot-pop','median-income','no-housing-units','med-home-val','owner-occupied','house-w-debt','house-wo-debt']})
for i in bigdata:  # create and print an array of listed classes
    print(i.string)
```

```python
print(bigdata[0].string) # array follows file's class order
print(bigdata[1].string)
```
